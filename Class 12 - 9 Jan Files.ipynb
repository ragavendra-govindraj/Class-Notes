```python
def gencube(n):
    l = []
    for i in range(n):
        l.append(i**3)
    return l
    
```

```python
gencube(60000)
```

```python
def gencube(n):
    for i in range(n):
        yield i**3
```

```python
gencube(6)
```

```python
# yield is a generator function
```

```python
for i in gencube(9):
    print(i)
```

```python
def fib(n):
    a = 1
    b = 1
    for i in range(n):
        yield a,i
        a,b = b, a+b
```

```python
for i in fib(10):
    print(i)
```

```python
def fib1(n):
    a = 1
    b = 1
    output = []
    for i in range(n):
        output.append(a)
        a,b = b, a+b
        
    return output

fib1(10)
```
